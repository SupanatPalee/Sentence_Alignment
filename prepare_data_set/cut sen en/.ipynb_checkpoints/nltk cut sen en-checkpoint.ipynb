```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
import os

# ตรวจสอบว่ามีโฟลเดอร์ nltk_data ถ้ายังไม่มี ให้สร้างขึ้นมา
nltk_data_folder = os.path.join(os.path.expanduser('~'), 'nltk_data')
if not os.path.exists(nltk_data_folder):
    os.makedirs(nltk_data_folder)

# ดาวน์โหลดตัวตัดประโยคของ NLTK
nltk.download('punkt', download_dir=nltk_data_folder)

# ฟังก์ชันสำหรับตัดประโยค
def tokenize_sentences(text):
    return sent_tokenize(text)

# วนลูปอ่านไฟล์ CSV
num_files = 1693  # จำนวนไฟล์ที่ต้องการประมวลผล
input_folder = r'C:\Users\supan\webscrap_v2\webscrap_v2\data\sort'  # แก้ไขให้เป็นที่อยู่โฟลเดอร์ไฟล์ input ของคุณ
output_folder = r'C:\Users\supan\webscrap_v2\webscrap_v2\cut sen en\output cut sentence en'  # แก้ไขให้เป็นที่อยู่โฟลเดอร์ไฟล์ output ของคุณ

# ตรวจสอบว่ามีโฟลเดอร์ output_folder ถ้ายังไม่มี ให้สร้างขึ้นมา
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for i in range(num_files):
    input_file_path = os.path.join(input_folder, f'{i}.csv')
    output_file_path = os.path.join(output_folder, f'{i}_sentences.csv')
    
    # ตรวจสอบว่าไฟล์ input มีอยู่จริง
    if not os.path.exists(input_file_path):
        print(f'File {input_file_path} does not exist. Skipping...')
        continue

    # อ่านไฟล์ CSV
    df = pd.read_csv(input_file_path)

    # ตรวจสอบคอลัมน์ของ DataFrame
    print(df.columns)

    # สมมติว่าคอลัมน์ที่ต้องการตัดประโยคคือ 'English_sentence'
    text_column = 'English_sentence'  # แก้ไขให้เป็นชื่อคอลัมน์ที่ต้องการตัดประโยค

    # ใช้ฟังก์ชันตัดประโยคกับแต่ละแถวในคอลัมน์
    df['sentences'] = df[text_column].apply(tokenize_sentences)

    # บันทึกผลลัพธ์เฉพาะคอลัมน์ 'sentences' ลงไฟล์ CSV ใหม่
    sentences_df = df[['sentences']].explode('sentences')
    sentences_df.to_csv(output_file_path, index=False, header=False)

    print(f'Processed and saved: {output_file_path}')

print('Completed processing all files.')
```
