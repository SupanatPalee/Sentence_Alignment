```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
import os

#  nltk_data  
nltk_data_folder = os.path.join(os.path.expanduser('~'), 'nltk_data')
if not os.path.exists(nltk_data_folder):
    os.makedirs(nltk_data_folder)

#  NLTK
nltk.download('punkt', download_dir=nltk_data_folder)

# 
def tokenize_sentences(text):
    return sent_tokenize(text)

#  CSV
num_files = 1693  # 
input_folder = r'C:\Users\supan\webscrap_v2\webscrap_v2\data\sort'  #  input 
output_folder = r'C:\Users\supan\webscrap_v2\webscrap_v2\cut sen en\output cut sentence en'  #  output 

#  output_folder  
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for i in range(num_files):
    input_file_path = os.path.join(input_folder, f'{i}.csv')
    output_file_path = os.path.join(output_folder, f'{i}_sentences.csv')
    
    #  input 
    if not os.path.exists(input_file_path):
        print(f'File {input_file_path} does not exist. Skipping...')
        continue

    #  CSV
    df = pd.read_csv(input_file_path)

    #  DataFrame
    print(df.columns)

    #  'English_sentence'
    text_column = 'English_sentence'  # 

    # 
    df['sentences'] = df[text_column].apply(tokenize_sentences)

    #  'sentences'  CSV 
    sentences_df = df[['sentences']].explode('sentences')
    sentences_df.to_csv(output_file_path, index=False, header=False)

    print(f'Processed and saved: {output_file_path}')

print('Completed processing all files.')
```
